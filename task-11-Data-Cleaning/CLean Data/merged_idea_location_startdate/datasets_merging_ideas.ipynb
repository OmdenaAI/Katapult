```python
import pandas as pd
import numpy as np
```

# Merging datasets that contains the Idea of the startup

## Cleaning LinkedIn dataset 

```python
df_linkedin = pd.read_csv("../../Data/Linkedin EM exited companies.csv")
df_linkedin.head(5)
```

```python
df_linkedin_cleaned = pd.DataFrame(df_linkedin[['Company Name', 'Overview', 'Description', 'Founded Year', 'City', 'State', 'Country', 'Company Stage', 'Acquired Year', 'Website Status', ]])
df_linkedin_cleaned
```

```python
df_linkedin_cleaned['Description'] = df_linkedin_cleaned.apply(lambda x: x['Overview'] if pd.isna(x['Description']) else x['Description'], axis=1)
df_linkedin_cleaned['City'] = df_linkedin_cleaned['City'] + " " + df_linkedin_cleaned['Country']
df_linkedin_cleaned['operating_years'] = df_linkedin_cleaned['Acquired Year'] - df_linkedin_cleaned['Founded Year']
df_linkedin_cleaned['number_employees'] = np.nan
df_linkedin_cleaned['latest_funding_round'] = np.nan
df_linkedin_cleaned.drop(columns = ['Acquired Year', 'Overview', 'State', 'Country'], inplace = True)
df_linkedin_cleaned.rename(columns={
    'Company Name': 'name',
    'Description': 'idea',
    'Founded Year': 'start_date',
    'City': 'location',
    'Company Stage': 'status',
    'Website Status': 'website_status'}, inplace=True)
df_linkedin_cleaned
```

## Cleaning startuptracker.io dataset 

```python
df_startuptracker = pd.read_csv("../../../Task 4 - Data Collection/start_up_io/data_from_startup_io.csv")
df_startuptracker.head(5)
```

```python
df_startuptracker_cleaned = pd.DataFrame(df_startuptracker[['name', 'description', 'founded', 'place', 'team_member', 'activity_verified' ]])
df_startuptracker_cleaned
```

```python
df_startuptracker_cleaned.drop(df_startuptracker_cleaned.tail(1).index, inplace=True)
df_startuptracker_cleaned['founded'] = df_startuptracker_cleaned['founded'].apply(lambda x: x[-4:])
df_startuptracker_cleaned['activity_verified'] = '20' + df_startuptracker_cleaned['activity_verified'].apply(lambda x: x[-2:])
df_startuptracker_cleaned['operating_years'] = df_startuptracker_cleaned['activity_verified'].astype(int) - df_startuptracker_cleaned['founded'].astype(int)
df_startuptracker_cleaned['status'] = np.nan
df_startuptracker_cleaned['latest_funding_round'] = np.nan
df_startuptracker_cleaned['website_status'] = np.nan
df_startuptracker_cleaned.drop(columns = ['activity_verified'], inplace = True)
df_startuptracker_cleaned.rename(columns={
    'description': 'idea',
    'founded': 'start_date',
    'place': 'location',
    'team_member': 'number_employees'}, inplace=True)
df_startuptracker_cleaned
```

## Cleaning autopsy_io (failed startups) dataset 

```python
df_autopsy = pd.read_csv("../../../Task 4 - Data Collection/Datasets for Baseline Model/autopsy_io/failed_companies_autopsyio_dataset.csv")
df_autopsy.head(5)
```

```python
df_autopsy_cleaned = pd.DataFrame(df_autopsy[['Startup', 'Idea', 'Startup_link' ]])
df_autopsy_cleaned
```

```python
# the "Startup_link" contains information of the startup that can help fill other missing features! like location, number of employees, etc.
df_autopsy_cleaned['start_date'] = np.nan
df_autopsy_cleaned['location'] = np.nan
df_autopsy_cleaned['status'] = 'Failed'
df_autopsy_cleaned['number_employees'] = np.nan
df_autopsy_cleaned['operating_years'] = np.nan
df_autopsy_cleaned['latest_funding_round'] = np.nan
df_autopsy_cleaned['website_status'] = np.nan
df_autopsy_cleaned.drop(columns = ['Startup_link'], inplace = True)
df_autopsy_cleaned.rename(columns={
    'Startup': 'name',
    'Idea': 'idea'}, inplace=True)
df_autopsy_cleaned
```

## Cleaning Failory's Startup Cemetery (failed startups) dataset 

```python
df_cemetery = pd.read_csv("../../../Task 4 - Data Collection/Datasets for Baseline Model/failory_startup_cemetery_data/failory_cemetery_failedstartups.csv")
df_cemetery.head(5)
```

```python
df_cemetery_cleaned = pd.DataFrame(df_cemetery[['failed_startup', 'full_description', 'year_start', 'country', 'year_closed', 'outcome', 'num_employees', 'num_funding_rounds', ]])
df_cemetery_cleaned
```

```python
#df_cemetery_cleaned = pd.DataFrame(df_cemetery_cleaned[~df_cemetery_cleaned['num_employees'].str.contains('[a-z]')])
df_cemetery_cleaned['year_closed'] = df_cemetery_cleaned.apply(lambda x: '2020' if x['year_closed'] == 'Active' else x['year_closed'], axis=1)
df_cemetery_cleaned['operating_years'] = df_cemetery_cleaned['year_closed'].astype(int) - df_cemetery_cleaned['year_start'].astype(int)
df_cemetery_cleaned['website_status'] = np.nan
df_cemetery_cleaned.drop(columns = ['year_closed'], inplace = True)
df_cemetery_cleaned.rename(columns={
    'failed_startup': 'name',
    'full_description': 'idea',
    'year_start': 'start_date',
    'country': 'location',
    'outcome': 'status',
    'num_employees': 'number_employees',
    'num_funding_rounds': 'latest_funding_round'
}, inplace=True)
df_cemetery_cleaned
```

## Cleaning 808 companies dataset

```python
df_808companies = pd.read_csv("../../../Task 4 - Data Collection/AngelList/808 companies_2018.csv")
df_808companies.head(5)
```

```python
df_808companies_cleaned = pd.DataFrame(df_808companies[['name', 'desc', 'location', 'employees' ]])
df_808companies_cleaned
```

```python
df_808companies_cleaned = pd.DataFrame(df_808companies_cleaned[(df_808companies_cleaned['employees'] != '-') | (df_808companies_cleaned['employees'].isna())])
df_808companies_cleaned = pd.DataFrame(df_808companies_cleaned[(df_808companies_cleaned['desc'] != '-') | (df_808companies_cleaned['desc'].isna())])
df_808companies_cleaned['start_date'] = np.nan
df_808companies_cleaned['status'] = np.nan
df_808companies_cleaned['latest_funding_round'] = np.nan
df_808companies_cleaned['operating_years'] = np.nan
df_808companies_cleaned['website_status'] = np.nan
df_808companies_cleaned.rename(columns={
    'desc': 'idea',
    'employees': 'number_employees'}, inplace=True)
df_808companies_cleaned
```

# Stacking all datasets

```python
print(sorted(df_linkedin_cleaned.columns))
print(sorted(df_startuptracker_cleaned.columns))
print(sorted(df_autopsy_cleaned.columns))
print(sorted(df_cemetery_cleaned.columns))
print(sorted(df_808companies_cleaned.columns))
print(len(df_linkedin_cleaned))
print(len(df_startuptracker_cleaned))
print(len(df_autopsy_cleaned))
print(len(df_cemetery_cleaned))
print(len(df_808companies_cleaned))


COLUMNS = ['name', 'idea', 'location', 'start_date', 'operating_years', 'number_employees', 'status', 'latest_funding_round']
```

```python
df_final = pd.concat([df_linkedin_cleaned[COLUMNS],
                      df_startuptracker_cleaned[COLUMNS],
                      df_autopsy_cleaned[COLUMNS],
                      df_cemetery_cleaned[COLUMNS],
                      df_808companies_cleaned[COLUMNS]], axis = 0, ignore_index=True)
df_final
```

```python
# Dropping duplicates
df_final.drop_duplicates(subset=COLUMNS, inplace = True)
df_final
```

```python
# See if there are duplicates by name
df_final.drop_duplicates(subset='name')
```
