```python
def set_viewport_size(driver, width, height):
    window_size = driver.execute_script("""
        return [window.outerWidth - window.innerWidth + arguments[0],
          window.outerHeight - window.innerHeight + arguments[1]];
        """, width, height)
    driver.set_window_size(*window_size)
```

```python
options = Options()
options.add_argument("--start-maximized")
bot=webdriver.Chrome(ChromeDriverManager().install(),options=options)
bot.get("https://www.linkedin.com/uas/login")
time.sleep(1.5)
email=bot.find_element_by_id("username")
email.send_keys(e_mail)
password=bot.find_element_by_id("password")
password.send_keys(password_)
time.sleep(1.5)
password.send_keys(Keys.RETURN)
index= 0
df['Number of employees']= ['NaN' for item in range(df.shape[0])]
df['Number of followers']= ['NaN' for item in range(df.shape[0])]
for startup_name in df.Name:
    try:  
        number_result = 3
        startup_name= startup_name.replace(' ','%20')
        search_url= "https://www.linkedin.com/search/results/companies/?keywords="+startup_name+"&origin=GLOBAL_SEARCH_HEADER"
        bot.get(search_url)
    
        
        time.sleep(1.5)
        linkedin_url= bot.find_element_by_class_name('entity-result__title-text').find_element_by_class_name('app-aware-link').get_attribute('href')
        #print(linkedin_url)
        #bot.get(linkedin_url)
        
        bot.get(linkedin_url+'/about/')
        time.sleep(1.5)
        number_of_employees= bot.find_element_by_class_name('org-about-company-module__company-size-definition-text').text.split()[0]
        number_of_followers= int(bot.find_elements_by_class_name('org-top-card-summary-info-list__info-item')[-1].text.replace('\u202f','').split()[0])
        #print(number_of_employees,number_of_followers)
        df['Number of employees'][index]= number_of_employees
        df['Number of followers'][index]= number_of_followers
        index= index+1
    except: 
        continue 
    
```
